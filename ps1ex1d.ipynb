```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
def equation(x):
    f = 2*x+3*np.tan(x/4)-np.exp(x)
    return f         
```

```python
def bisection(a,b,error):
    erN = abs(b-a)/2
    n=0
    root = 0.8335555793437379
    #this is the root we calculated for error smaller than 10**-12
    x = []
    y1 = []
    y2 = []
    EN = []

    pre = [abs(root-((a+b)/2))]
                 
    #we already know there is a root for f and so we don't need to check again
    while erN>=error:
        c=(a+b)/2
        n = n+1
        d = equation(c)
        x.append(n)
        y1.append(d)
        y2.append(abs(root-c))
        f1 = equation(a)
        f2= equation(b)
        f3 = equation(c)
        s1 = f1*f3
        s2 = f2*f3
        if s1<0:
            b = c
            erN = abs(b-a)/2
            EN.append(abs(root-((b+a)/2)))
        if s2<0:
            a = c
            erN = abs(b-a)/2
            EN.append(abs(root-((b+a)/2)))
            

  
    #That was the main function which calculates the actual error
    #Now we will procede to calculate the predicted error
    

    for i in range (0,n-1):
        pre.append(y2[i]/2)
        print("\n",x[i],y2[i],pre[i],)
    
    #I'm not sure about this one. I tried defining pre = [] at the beggining
    #but the graph of the predicted error was moved left. And even though the numbers for many iterations complied 
    #the relation between y2 and pre was not correct 
    #because y2[0] gave us pre[0] which is supposedly equal to y2[1]
    #but y2[0] and pre[0]=y2[1] were shown over the same n
    #so y2[1] was compared to y2[0] and not y2[1]
    
    print("\nFor maximum error of", error, ", N is: ", n)
    print("When the f(xn)s are concered, the f approaches zero in the bellow manner:")
    graph1 = plt.plot(x,y1, marker = 'o')
    plt.show(graph1)
    print("And the various xns differ from the real root according to the graph:")
    graph2 = plt.plot(x,y2, marker = 'o')
    plt.show(graph2)
    graph3 = plt.plot(EN,y2, marker = 'o')
    print("And this is a graph of εn+1 in relation to εn")
    plt.show(graph3)

    print("\n And in a logarithmic scale we can present the evaluated and predicted error as:")
    fig, ax = plt.subplots()
    ax.plot(x, y2, x, pre, marker='o')
    ax.set_yscale('log')
    ax.set(xlabel='iteration', ylabel='error')
    ax.grid()
    plt.xlim(right=n)
    tick_spacing = 1
    ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
    plt.show()
      
    
                      
    #we can (instead of the print command) use a return command, giving us the option to manage the results and/or use them in later calculations    
```

```python
a = 0.5
b = 1
bisection(a,b,10**(-3))
bisection(a,b,10**(-6))
bisection(a,b,10**(-12))
```
