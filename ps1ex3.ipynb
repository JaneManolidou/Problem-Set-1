```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
def fun(x):
    f = 2*x+3*np.tan(x/4)-np.exp(x)
    return f  
def der(x):
    d = 2 + 3*(1/(4*((np.cos(x/4))**2)))- np.exp(x)
    return d
def der2(x):
    d = -np.exp(x)+(3/8)*np.tan(x/4)/(np.cos(x/4)**2)
    return d
```

```python
def newtraph(xo,N):
    n = 0
    f = fun(xo)
    d = der(xo)
    solution = [xo]
    x = xo - (f/d)
    while n < N:
        solution.append(x)
        print((n, x))
        #just to check because after a while we get 0 error
        n = n+1
        f = fun(x)
        d = der(x)
        x = x - (f/d)
    print("The solution is: ", x)
    #we know the number of iterations so let's calculate the error
    #and we can also calculate the predicted error
    er = []
    it = []
    pre = []
    for i in range (0,N-1):
        error = x - solution[i]
        it.append(i)
        er.append(error)
        k = -((error**2)/2)*(der2(x)/der(x))
        pre.append(k)
        
    
    
    
    
    print("And for the error we have:", error)
    fig, ax = plt.subplots()
    ax.plot(it, er,it,pre, marker='o')
  
    ax.set_yscale('log')
    ax.set(xlabel='iteration', ylabel='error')
    ax.grid()
    plt.xlim(right=N)
    tick_spacing = 1
    ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing)) 
    print("It approaches this value according to the graph:")
    plt.show()
```

```python
def newtraph2(xo,er):
    n = 0
    N = 0
    f = fun(xo)
    d = der(xo)
    x = xo - (f/d)
    while n == 0:
        f = fun(x)
        d = der(x)
        x2 = x - (f/d)
        error = abs(x-x2)
        x = x2
        N = N + 1
        if error < er:
            break
    print("The solution is:", x, "  with an error of:", error)
    print("It took us ", N, " iterations to reach this")
        
```

```python
xo = 0.5
print("Let's start with N = 8 iterations:")
newtraph(xo,8)
print("And proceed with 18:")
newtraph(xo,18)
print("And finally for 38:")
newtraph(xo,38)
```

```python
print("Now let us check with the error")
print("For error smaller than 10^(-3):")
newtraph2(xo,10**(-3))
print("For error smaller than 10^(-6):")
newtraph2(xo,10**(-6))
print("For error smaller than 10^(-12):")
newtraph2(xo,10**(-12))
```
